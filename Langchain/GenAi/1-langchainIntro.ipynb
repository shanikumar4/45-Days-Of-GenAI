### LLM Model Integration With Google Gemini

```python
os.environ["GOOGLE_API_KEY"]= os.getenv("GEMINI_API_KEY")
```

```python
from langchain.chat_models import init_chat_model
model = init_chat_model("google_genai:gemini-3.5-flash")
response= model.invoke("Write me an essay about Artificial Intellignece")
print(response.content)
```

```python
from langchain_google_genai import ChatGoogleGenerativeAI
model=ChatGoogleGenerativeAI(model="gemini-3.5-flash")
response= model.invoke("Write me an essay about Artificial Intellignece")
print(response.content)
```

```python
import os
from langchain.chat_models import init_chat_model
os.environ["GROQ_API_KEY"]=os.getenv("GROQ_API_KEY")

model = init_chat_model("groq:llama-3.1-8b-instant")
response= model.invoke("Write me a joke on machine learning algorithms")
print(response.content)

```
