Token


```python
import tiktoken
```

```python
encoding = tiktoken.get_encoding("cl100k_base")
```

```python
text = "Large Language Models are amazing!"

tokens = encoding.encode(text)

print("Tokens:", tokens)
print("Number of tokens:", len(tokens))
```

```python
decoded = encoding.decode(tokens)

print("Decoded:", decoded)
```

Context Window

```python
context_window = 10

conversation = [
    "Hi",
    "How are you?",
    "I'm learning AI.",
    "Today's topic is LLMs.",
    "Scaling Laws",
    "Transformers",
    "Attention",
    "Embeddings",
    "Temperature",
    "Top-p",
    "Goodbye"
]

visible = conversation[-context_window:]

print("Messages visible to the model:\n")

for msg in visible:
    print("-", msg)
```

Temperature

```python
from transformers import pipeline

generator = pipeline("text-generation", model="gpt2")

prompt = "Artificial Intelligence will"

result = generator(
    prompt,
    max_length=40,
    temperature=0.2
)

print(result[0]["generated_text"])
```

Top-p Sampling

```python
from transformers import pipeline

generator = pipeline("text-generation", model="gpt2")

output = generator(
    "AI will change the future because",
    max_length=50,
    temperature=0.8,
    top_p=0.9,
    do_sample=True
)

print(output[0]["generated_text"])
```
